## 17. What is the result of the following expression? (★☆☆)


## 19. Create a 8x8 matrix and fill it with a checkerboard pattern (★☆☆)

````python
A = np.zeros((8,8))
A[1::2,::2] = 1
A[::2,1::2] = 1
print(A)
````

## 22. Normalize a 5x5 random matrix (★☆☆)

````python
A = np.random.randint(10, size=(5,5))
m = np.mean(A)
s = np.std(A)
B = (A - m)/s
print(B)

max = np.max(A)
min = np.min(A)
C = (A - min) / (max - min)
print(C)
````
